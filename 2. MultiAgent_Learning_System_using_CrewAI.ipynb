## Creating Agents

- Define the Agents, and provide them a `role`, `goal` and `backstory`.


## 1. Subject Matter Researcher

```python
researcher = Agent(
    role="Content Planner",
    goal="Gather and summarize accurate, relevant, and well-structured information about the topic: {topic}.",
    backstory="You are a skilled researcher responsible for planning educational content on the topic: {topic}. "
              "Your task is to collect, and organize key information that helps learners gain a deep understanding of the subject. "
              "Your findings provide the foundation for the Content Writer to develop clear and effective educational materials. "
              "You focus on aligning your research with learning objectives and ensuring that all information is both accessible and factually sound.",
    allow_delegation=False,
    verbose=True
)
```

## 2. Learning Material Developer

```python
content_creator = Agent(
    role="Content Writer",
    goal="Write well-structured, factually accurate, and accessible content for learners on the topic: {topic}.",
    backstory="You are responsible for writing educational content on the topic: {topic}. "
              "You collaborate closely with the Content Planner, who provides a structured outline and key contextual information. "
              "Your job is to transform the outline into a coherent and engaging article that meets the learning objectives. "
              "You ensure the information is accurate, appropriately simplified for the target audience, and backed by reliable sources. "
              "Your tone is clear, supportive, and designed to maximize learner comprehension.",
    allow_delegation=False,
    verbose=True
)
```

## 3. Educational Content Editor

```python
edu_editor = Agent(
    role="Educational Content Editor",
    goal="Edit a given educational article to ensure it is clear, accurate, and aligned with the target learning level and instructional goals.",
    backstory="You are an experienced educational content editor who receives draft content from content writer. "
              "Your goal is to review the material to improve clarity, correct grammar, and structure it in a way that supports comprehension. "
              "You also ensure the tone is age-appropriate, the examples are relevant and effective, and the information aligns with pedagogical standards. "
              "You avoid jargon unless it is defined, and prioritize learner engagement and retention.",
    allow_delegation=False,
    verbose=True
)
```

## 4. Flashcard Generator

```python
flashcard_creator = Agent(
    role="Flashcard Generator",
    goal="Generate clear, concise flashcards from educational content to help learners revise and reinforce key concepts.",
    backstory="You specialize in transforming detailed educational material into effective revision tools. "
              "Your job is to extract essential facts, definitions, and concepts and turn them into digestible question-and-answer pairs. "
              "These flashcards should support active recall and spaced repetition, helping learners retain key information more effectively.",
    allow_delegation=False,
    verbose=True
)
```

## 5. Assessment Designer

```python
quiz_builder = Agent(
    role="Assessment Designer",
    goal="Create a quiz with multiple-choice questions that accurately assess learners’ understanding of the topic: {topic}.",
    backstory="You are an expert in educational assessment design. Your role is to craft clear, fair, and pedagogically sound multiple-choice questions "
              "based on the educational content. Each question should test key concepts, provide plausible distractors, and reinforce learning through challenge. "
              "Your quizzes are used to measure how well learners have grasped the material and to identify areas that may need review.",
    allow_delegation=False,
    verbose=True
)
```

# Creating Tasks

```python
from crewai import Task

# Task 1: Research
outline_task = Task(
    description=(
        "Conduct thorough research on the topic: {topic}.\n"
        "Break down the topic into teachable sections, aligned with educational outcomes."
    ),
    expected_output="A detailed research summary with core concepts, definitions, and examples.",
    agent=researcher
)


# Task 2: Create Learning Material
create_material_task = Task(
    description=(
        "Based on the research summary, create a well-structured learning module.\n"
        "Include an introduction, key points, and a short recap.\n"
        "Use headings and bullet points to improve readability."
    ),
    expected_output="An educational lesson in Markdown format with clear structure and learner-friendly tone.",
    agent=content_creator
)

# Task 3: Edit Content
edit_content_task = Task(
    description=(
        "Edit the educational content to ensure clarity, correctness, and consistent voice.\n"
        "Improve sentence structure, tone, and formatting for learners."
    ),
    expected_output="A polished and professionally edited Markdown lesson.",
    agent=edu_editor
)

# Task 4: Generate Flashcards
flashcard_task = Task(
    description=(
        "Create 5-10 flashcards from the finalized content.\n"
        "Each card should have a question on one side and the answer on the other."
    ),
    expected_output="A list of Q&A pairs suitable for flashcard-based revision.",
    agent=flashcard_creator
)

# Task 5: Build Quiz
quiz_task = Task(
    description=(
        "Create 5-7 multiple-choice questions to test the learner's understanding.\n"
        "Each question should have 4 options with only one correct answer."
    ),
    expected_output="A short quiz with 5 multiple-choice questions and answers.",
    agent=quiz_builder
)
```

# Create Crews

```python
from crewai import Crew

crew = Crew(
    agents=[
        researcher,
        content_creator,
        edu_editor,
        flashcard_creator,
        quiz_builder
    ],
    tasks=[
        outline_task,
        create_material_task,
        edit_content_task,
        flashcard_task,
        quiz_task
    ],
    verbose=2
)

# Run the crew with a topic
result = crew.kickoff(inputs={"topic": "Compound AI Agents"})

```
